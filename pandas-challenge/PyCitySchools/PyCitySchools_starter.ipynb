```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of schools
#Calculate the total number of students
#Calculate the total budget
#Calculate the average math score
#Calculate the average reading score
totSchool = len(School_df["School ID"].unique())
totStudent = len(Student_df["Student ID"].unique())
totBudget = School_df["budget"].sum()
AvgMathScoreStu = Student_df["math_score"].mean()
AvgReadingScoreStu = Student_df["reading_score"].mean()
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
#Calculate the percentage of students with a passing reading score (70 or greater)

PerReadingPass = (Student_df.query("reading_score>=70")["Student ID"].count()/totStudent) * 100
PerMathPass= (Student_df.query("math_score>=70")["Student ID"].count()/totStudent) * 100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
PerMath_ReadPass = (Student_df.query("math_score>=70 and reading_score>=70")["Student ID"].count()/totStudent) * 100
```

```python
#Summary dataframe
districtSummary_df = pd.DataFrame({
    "Total Schools":[totSchool], 
    "Total Students":[totStudent],
    "Total Budget":[totBudget],
    "Average Math Score":[AvgMathScoreStu],
    "Average Reading Score":[AvgReadingScoreStu],
    "% Passing Math":[PerMathPass],
    "% Passing Reading":[PerReadingPass],
    "% Overall Passing":[PerMath_ReadPass]
})
```

```python
## format data values
districtSummary_df["Total Budget"] = districtSummary_df["Total Budget"].map("${:,.2f}".format)
districtSummary_df["Total Students"] = districtSummary_df["Total Students"].map("{:,}".format) 

##summary dataframe
districtSummary_df.head()
```
